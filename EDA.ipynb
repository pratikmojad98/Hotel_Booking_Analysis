### Variables Description

**hotel:**  Resort Hotel or City Hotel

**arrival_date_week_number:** week number of arrival date

**arrival_date_day_of_month:**  Day of arrival date

**stays_in_weekend_nights:**  Number of weekend nights the guest stayed or booked to stay at the hotel

**children:**  Number of children

**babies:**  Number of babies

**meal:** Meal types- BB(Bed and breakfast) , FB(Full board) , HB(Half board) , SC()

**country:**  Country code

**market_segment:**  Which segment customer belong to: [Direct,Corporate,Online TA,Offline TA/TO,Complementary,Groups,Undefined,Aviation]

**distribution_channel:**  How the customer accessed the stay:[Direct,Corporate,TA/TO,Undefined,GDS]

**is_repeated_guest:**  0(If guest coming for first time) and 1 (If repeated guest)

**previous_cancellations:** was there a cancellation before

**previous_bookings_not_canceled:** Count of previous booking that was not canceled

**reserved_room_type:**  Reserved room type

**assigned_room_type:**  Assigned room type

**booking_changes:** Count of changes made to booking

**deposit_type:**  Type of deposit : [No Deposit,Refundable,Non Refund]

**agent:** Booked through agent

**company:**  Company

**days_in_waiting_list:**  Number of days in waiting list

**customer_type:**  Type of customer: [Transient, Contract,Transient-Party,Group]

**adr:** Average Daily Rate of hotel

**required_car_parking_spaces:**  Count of required car parking spaces

**total_of_special_requests:**  Number of additional requests

**reservation_status:**  Reservation Status : [Check-Out,Canceled,No-Show]

**reservation_status_date:**  Date of the specific status

````python
# Defining function which returns date in sorted format YYYY-MM-DD.
def get_date(year,month,day):
  d = '-'.join([str(year),month[:3],str(day)])
  return d
````

````python
#Creating new column 'arrival date' having date in format YYYY-MM-DD.
df['arrival_date'] = df.apply(lambda x: get_date(x['arrival_date_year'],x['arrival_date_month'],x['arrival_date_day_of_month']), axis=1)
df['arrival_date'] = pd.to_datetime(df['arrival_date'])
````

````python
## Creating new column 'df_weekend_resort_hotel_1' which shows  Number of weekend nights the guest stayed at resort hotel and their Frequency.
df_weekend_resort_hotel = df.groupby("hotel",as_index=False)["stays_in_weekend_nights"].get_group('Resort Hotel').reset_index()
df_weekend_resort_hotel = df_weekend_resort_hotel.drop(['index'], axis=1)
df_weekend_resort_hotel_1 = df_weekend_resort_hotel.value_counts().reset_index()
df_weekend_resort_hotel_1.columns = ['stays_in_weekend_nights','Frequency']
df_weekend_resort_hotel_1
````
